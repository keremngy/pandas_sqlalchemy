```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('depremsenaryosu.csv', sep=';',encoding='utf-8')

df.head(3)
```

```python
df.rename(columns=lambda x: x.lower().replace(' ','_'),inplace=True)

df.head(3)
```

```python
df.info()
```

```python
df.drop(df.columns[[2,11,12,13,14]],axis=1, inplace=True)

df.head(3)
```

```python
df = df.rename(columns={"i̇lce_adi":"ilce",
                        "mahalle_adi":"mahalle",
                        "cok_agir_hasarli_bina_sayisi":"cokhasbinasay",
                        "agir_hasarli_bina_sayisi" : "agirhasbinsay",
                        "orta_hasarli_bina_sayisi" : "ortahasbinsay",
                        "hafif_hasarli_bina_sayisi" : "hafifhasbinsay",
                        "can_kaybı_sayisi" : "cankaysay",
                        "agir_yarali_sayisi" : "agiryarsay",
                        "hastanede_tedavi_sayisi" : "hastedsay",
                        "hafif_yarali_sayisi" : "hafifyarsay"
                       })

df.head(2)
```

```python
df.isnull().sum()
```

```python
dfilce = df.groupby("ilce").sum()

dfilce.head(3)
```

```python
toplambinalar = dfilce['cokhasbinasay'] + dfilce['agirhasbinsay']+dfilce['ortahasbinsay']+dfilce['hafifhasbinsay']

dfilce['toplambinasayisi'] = toplambinalar
```

```python
toplamyaralisayisi = dfilce['agiryarsay']+dfilce['hastedsay']+dfilce['hafifyarsay']

dfilce['toplamyaralisayisi'] = toplamyaralisayisi
```

```python
cokveagirhasbinalar = dfilce['cokhasbinasay'] + dfilce['agirhasbinsay']

dfilce['bindenfazlaetkbinalar'] = np.where(cokveagirhasbinalar > 1000, 'Yüksek','Düşük')
```

```python
cankaybisayisiortustunde = dfilce['cankaysay'] > dfilce['cankaysay'].mean()

dfilce['olumsayisiortgore'] = np.where(cankaybisayisiortustunde,'Yüksek','Düşük' )
```

```python
agirvehastaneyaralisay = dfilce['agiryarsay']+dfilce['hastedsay']

durumlar = [
    (agirvehastaneyaralisay >= 5000) & (agirvehastaneyaralisay < 10000),
    (agirvehastaneyaralisay >= 3000) & (agirvehastaneyaralisay < 5000),
    (agirvehastaneyaralisay >= 1000) & (agirvehastaneyaralisay < 3000), 
    (agirvehastaneyaralisay >= 500) & (agirvehastaneyaralisay < 1000),
    (agirvehastaneyaralisay >= 1) & (agirvehastaneyaralisay <= 500),
    (agirvehastaneyaralisay == 0)
           ]

degerler = ['5000-10000 Kişi','3000-5000 Kişi','1000-3000 Kişi','500-1000 Kişi','100-500 Kişi','Yok']

dfilce['oncelikliyaralidurumu'] = np.select(durumlar,degerler)
```
